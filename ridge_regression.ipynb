```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, Lasso, Ridge, RidgeCV, LassoCV
from sklearn.model_selection import train_test_split, RepeatedKFold

import warnings

# Filter out import warnings
warnings.filterwarnings("ignore", category=ImportWarning)
```

```python
df = pd.read_csv('data/updated_data.csv')
```

```python
data = df.reindex(columns=['DEP_DELAY', 'CRS_ELAPSED_TIME', 'DISTANCE', 'CRS_DEP_M', 'DEP_TIME_M', 'CRS_ARR_M', 'Temperature', 'Dew_Point', 'Wind Gust', 'Pressure', 'sch_dep', 'sch_arr', 'TAXI_OUT'])
```

```python
data_to_use = data.iloc[:,0:12]
data_to_target = data.iloc[:,12]
```

```python
model_ridge = Ridge(alpha=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_to_use, data_to_target, test_size=20, random_state=32)
```

```python
model_ridge.fit(X_train,y_train)
```

```python
pred_ridge = model_ridge.predict(X_test)
pred_ridge
```

```python
## mnake prediction in a dataframe
ridge_df = pd.DataFrame({'Actual_score': y_test, 'Predicted_score': pred_ridge})
ridge_df
```

```python
#tuning alpha
#n_repeats number of times of training it will shuffle and repeat ...computeration and space needs to be consider
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

model = RidgeCV(alphas=(0.5, 1, 0.01), cv=cv, scoring='neg_mean_absolute_error')

## fit the model
model.fit(X_train,y_train)
```
